```python
movie_name = []
year = []
runtime = []
imdbrate = []
metascore = []
rating = []
director = []
counter = 0

for page in range(1, 10000, 50): 
#for page in range(1):

    URL = f'https://www.imdb.com/search/title/?title_type=feature&genres=horror&start={page}&ref_=adv_nxt'
    response = requests.get(URL)
    soup = BS(response.text)
    tree = html.fromstring(response.content)
    counter += 1
    print(counter)
    for n in soup.findAll('div', attrs= {'class' : 'lister-item mode-advanced'}):
        movie_name.append(n.h3.a.text)
        try:
            for r in n.find(attrs= {'class' : 'runtime'}):
                if r:
                    runtime.append(r.get_text(strip=True).replace(' min', ""))
        except:
            runtime.append("N/A")
            
    for y in soup.findAll(attrs= {'class' : 'lister-item-year text-muted unbold'}):
        yr = y.get_text(strip=True)
        print(yr)
        print(type(str(yr)))
        yearclean = re.findall('[0-9]+', str(yr))
        yearclean = " ".join(yearclean)
        year.append(yearclean)
        #year.append(re.findall('[0-9]+', str(yr)))
        #year.append(re.findall('[0-9]+', str(yr))[0])
            
    for event in soup.find_all('div', {'class': 'lister-item-content'}):
        count = len(event.find_all('p'))
        imdbrating = event.find('div', {'class': 'inline-block ratings-imdb-rating'})
        if count > 3:
            imdbrate.append(imdbrating.get_text(strip=True))
        else:
            imdbrate.append("NO IMDB RATING")
            
    for c in soup.find_all('p', {'class': 'text-muted'}):
        count1 = len(c.find_all('span'))
        #print(count1)
        try:
            rating1 = c.find(attrs= {'class' : 'certificate'})
            if count1 == 0:
                continue
            elif count1 >= 4:
                rating.append(rating1.get_text(strip=True))
            else:
                rating.append("NO RATING")
        except:
            rating.append("NO RATING")   
            
    for m in soup.find_all('div', {'class': 'lister-item-content'}):
        count3 = len(m.find_all('div'))
        try:
            metascore1 = m.find('div', {'class' : 'inline-block ratings-metascore'})
            if metascore1:
                if count3 >= 1:
                    metascore.append(metascore1.get_text(strip=True).replace('Metascore', ''))
                else:
                    metascore.append("NO METASCORE")
            else:
                metascore.append("NO METASCORE")
        except:
            metascore.append("NO METASCORE")

                 
#     for y in soup.findAll(attrs= {'class' : 'lister-item-year text-muted unbold'}):
#             year.append(y.get_text(strip=True))
        
#     for r in soup.findAll(attrs= {'class' : 'runtime'}):
#             runtime.append(r.get_text(strip=True))
       
#     for i in soup.findAll("div", attrs= {'class' : 'inline-block ratings-imdb-rating'}):
#         imdbrate.append(i.get_text(strip=True))
        
#     for m in soup.findAll("div", attrs= {'class' : 'inline-block ratings-metascore'}):
#         metascore.append(m.get_text(strip=True))
        
#     for c in soup.findAll(attrs= {'class' : 'certificate'}):
#         rating.append(c.get_text(strip=True))
        
#     for d in soup.findAll('p', attrs= {'a' : 'href'}):
#          director.append(d.get_text(strip=True))
        
#WORKING ON DIRECTOR

#KEEP BUILDING LOOPS UNDER HERE

# //*[@id="main"]/div/div[3]/div/div[1]/div[3]/p[3]/a[1] - hreg tags for directors

# //*[@id="main"]/div/div[3]/div/div[2]/div[3]/p[3]/a[1]
```
